```python
from pathlib import Path

import pandas as pd

from test.architecture.commons import barplot_metric
```

```python
evaluation_results = Path('..') / 'datasets'
```

### Architecture Evaluation

```python
architecture_eval_df = pd.read_json(
    str(
        evaluation_results /
        'default_architecture_out.json'
    )
)

architecture_eval_df
```

```python
conversation_relevancy_results: dict[str, list[float]] = {model: [] for model in set(list(architecture_eval_df['model']))}
i = 0
for _, architecture_eval_row in architecture_eval_df.iterrows():
    conversation_relevancy_results[architecture_eval_row['model']].append(architecture_eval_row['conversation_relevancy'])
barplot_metric('conversation_relevancy', conversation_relevancy_results)
```

```python
conversation_completeness_results: dict[str, list[float]] = {model: [] for model in set(list(architecture_eval_df['model']))}
i = 0
for _, architecture_eval_row in architecture_eval_df.iterrows():
    conversation_completeness_results[architecture_eval_row['model']].append(architecture_eval_row['conversation_completeness'])
barplot_metric('conversation_completeness', conversation_completeness_results)
```
